# Customer List 

## Import packages

```python
import pandas as pd
import numpy as np
```

## Import raw data as raw_data


```python
raw_data = pd.read_excel("Customer Call List.xlsx")
raw_data
```

## Copy data as 'df' and show information about 'df'

```python
df = raw_data.copy()
df
```

```python
# Info about 'df'
df.info()
```

## Find and drop dupplicated values

```python
# Display duplicated values
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
df
```

```python
## Drop column "Not_useful_Column"
df.drop(["Not_Useful_Column"], axis = 1, inplace = True)
```

```python
df
```

## Clean df["Last_Name"]

```python
# Clean Last Name
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

## Clean data on Paying Customer and Do_Not_Contact

```python
# Change Y to Yes, N to No
df["Paying Customer"].unique()
```

```python
df["Paying Customer"].replace({"N": "No", "Y" : "Yes"}, inplace=True)
df["Do_Not_Contact"].replace({"N": "No", "Y" : "Yes"}, inplace=True)
```

```python
df
```

## Format "Phone_Number"

```python
df
```

```python
# Remove -,/,| from the phone numbre
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
# Transform into str
df["Phone_Number"] = df["Phone_Number"].apply(lambda x:str(x))
# Transform phone number
df["Phone_Number"] = df["Phone_Number"].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"].replace({"nan--" : "", "Na--":""}, inplace=True)
```

```python
df
```

```python
# Replace N/a witn blank
df.replace({"N/a" : ""}, inplace=True)
```

```python
df
```

## Recup Street Adreess, State and Zip_code

```python
# Recup Street Adreess, State and Zip_code
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
```

# Fill Na

```python
df = df.fillna('')
```
