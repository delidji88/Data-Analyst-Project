# Customer

## Import packages

```python
import pandas as pd
import numpy as np
```

## Import raw data as raw_data

```python
raw_data = pd.read_excel("Customer_data.xlsx")
raw_data
```

## Copy raw_data into df and display info about data

```python
# Copy raw_data into 'df'
df = raw_data.copy()
```

```python
# Display info about df
df.info()
```

## Drop certains columns

```python
# Drop columns 'driver_license', 'locality', 'address2'
df.drop(["address2", "driver_license", "locality","postal_code_plus4", "order_date", "order_posted_date",], axis=1, inplace = True)
```

```python
df.info()
```

## Drop Dupplicates and Nan values

```python
# Display dupplicates 
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Display nan
df.isna().value_counts()
```

```python
# Drop nan
df = df.dropna(how="any")
df
```

# Separate firstname and lastname

```python
df["firstname"] = df["cust_name"].apply(lambda x:x.split(" ")[0])
df["lastname"] = df["cust_name"].apply(lambda x:x.split(" ")[1])
df
```

```python
## Rename a column
df.rename(columns={"gender_code": "gender"}, inplace=True)
## Drop cust_name
df.drop(["cust_name"], axis = 1, inplace=True)
df
```

```python
df.info()
```

## Credit card

```python
df["creditcard_type"].unique()
```

```python
df["creditcard_type"].value_counts()
```

```python
# Replace American Expres by "American Express" or by using map
df["creditcard_type"].replace({"American Expres" : "American Express"}, inplace = True)
df["creditcard_type"].replace({"jcb" : "JCB"}, inplace = True)
df
```

```python
df.info()
```

## Recup Year, Month and Day

```python
#Year
df["Year"] = df["order_ship_date"].dt.year
#Month
df["Month"] = df["order_ship_date"].dt.month
#Day
df["Day"] = df["order_ship_date"].dt.day
df
```
