```python
import pandas as pd
url = "https://fr.wikipedia.org/wiki/Coupe_du_monde_de_football_2022"
wc2022 = pd.read_html(url)
wc2022
```

```python
goal_scorers = wc2022[87]
goal_scorers
```

```python
goal_scorers.columns
```

```python
goal_scorers.drop(['Place'], axis = 1, inplace = True)
goal_scorers
```

```python
goal_scorers["But par minutes"] = goal_scorers["Temps de jeu"] / goal_scorers["Buts"]
goal_scorers
```

```python
goal_scorers.sort_values(by=["But par minutes"], ascending=False)
```

```python
url2 = "https://fr.wikipedia.org/wiki/Transfert_(football)"
transfert = pd.read_html(url2)
transfert
```

```python
df = transfert[0]
df
```

```python
df[["Prix","Millions"]] = df['Transfert (en euros)'].str.split(' ',expand=True)
df
```

```python
df['Prix'] = df["Prix"].str.replace(",", ".")
df
```

```python
df.drop(["Millions"], axis = 1, inplace = True)
df
```

```python
df.info()
```

```python
df["Prix"] = df["Prix"].astype(float)
df.head()
```

```python
df.drop(['Transfert (en euros)'],axis=1,inplace=True)
df
```

```python
df = df.rename(columns={"Prix" : "Prix en millions d'Euro"})
df
```

```python
group_by = df[['Destination', "Prix en millions d'Euro"]].groupby("Destination").sum()
group_by
```

```python
group_by.sort_values(by=["Prix en millions d'Euro"], ascending = False)
```
