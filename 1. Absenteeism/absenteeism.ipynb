## Drop 'ID'

```python
df.drop(["ID"], axis=1, inplace = True)
df
```

## Change Date Dtype and recup year, month and date

```python
# Change Date Dtype
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
df
```

```python
# Recup year, month and day into df["Year"],
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df
```

## Education

```python
df["Education"].value_counts()
```

```python
# Change 1 to 0, 3 to 1, 2 to 1 and 4 to 1
df["Education"] = df["Education"].map({1:0,3:1,2:1,4:1})
```

```python
df["Education"].value_counts()
```

```python
df
```

## Work on Reason for Absence

```python
# Display Raison for Absence
sorted(df["Reason for Absence"].unique())
```

```python
# Create reason
reason = pd.get_dummies(df["Reason for Absence"], dtype = "int")
reason
```

```python
reason["Check"] = reason.sum(axis=1)
reason
```

```python
reason["Check"].sum(axis=0)
```

```python
reason.drop(["Check"],axis=1, inplace=True)
reason
```

```python
# Put together
reason_type_1 = reason.loc[:,0:14].max(axis=1)
reason_type_2 = reason.loc[:,15:17].max(axis=1)
reason_type_3 = reason.loc[:,18:21].max(axis=1)
reason_type_4 = reason.loc[:,22:].max(axis=1)
```

```python
#Concat
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
```

```python
# rename
df.columns.values
```
