# Sales project

## Import packages

```python
import numpy as np
import pandas as pd
import os
```

## Merge data

```python
# put all months data into on DataFrame called "all_months_data.csv"
path = './Sales_Data'
files = [file for file in os.listdir(path) if not file.startswith('.')]
all_months = pd.DataFrame()

for file in files :
    current_month = pd.read_csv(path+"/"+file)
    all_months = pd.concat([all_months, current_month])

all_months.to_csv("all_months_data.csv", index= False)
```

```python
# Display "all_months_data.csv" as raw_data
raw_data = pd.read_csv("all_months_data.csv")
raw_data
```

```python
# Copy data as 'df'
df = raw_data.copy()
df
```

```python
#Display informations about 'df'
df.info()
```

## Remove nan

```python
# Display nan
df.isna().value_counts()
```

```python
# Remove nan
df.dropna(how="all", inplace=True)
```

```python
df
```

## Remove dupplicate

```python
# Display dupplicate
df.duplicated().value_counts()
```

```python
# Drop dupplicates
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
# Display ligne 254
df.loc[254,:]
```

```python
# Drop ligne 254
df.drop([254,1], inplace=True)
```

```python
df
```

## Change Dtype of columns

```python
df.info()
```

```python
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])
df["Order Date"] = pd.to_datetime(df["Order Date"])
df
```

## Recup Year, Month and Day

```python
# Recup Year, Month and Day
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Day"] = df["Order Date"].dt.day
df["Hour"] = df["Order Date"].dt.hour
df["Minute"] = df["Order Date"].dt.minute
df
```

## Recup City and State

```python
# Recup City
df["City"] = df["Purchase Address"].apply(lambda x: x.split(",")[1])
df
```

```python
# Recup State
df["State"] = df["Purchase Address"].apply(lambda x: x.split(",")[2].split(" ")[1])
df
```

## Add Sales column

```python
# Create Sales column
df["Sales"] = df["Quantity Ordered"] * df["Price Each"]
df
```

## Reorganize columns

```python
# name of columns
df.columns.values
```

```python
# List of columns names
reorganized = ['Order ID', 'Product', 'Quantity Ordered', 'Price Each', 'Sales', 'Order Date', 'Year', 'Month', 'Day', 'Hour','Minute' , 'Purchase Address', 'City', 'State']
```

```python
# reoranization
df = df[reorganized]
```

```python
df
```

## Save cleaned data

```python
## Save cleaned data as Sales_Data_Cleaned
df.to_csv("Sales_Cleaned_data.csv", index= False)
```
