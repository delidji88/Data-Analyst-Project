## Import packages

```python
import numpy as np
import pandas as pd
import os
```

## Merge data

```python
# put all months data into on DataFrame called "all_months_data.csv"
path = './Sales_Data'
files = [file for file in os.listdir(path) if not file.startswith('.')]
all_months = pd.DataFrame()

for file in files :
    current_month = pd.read_csv(path+"/"+file)
    all_months = pd.concat([all_months, current_month])

all_months.to_csv("all_months_data.csv", index= False)
```
